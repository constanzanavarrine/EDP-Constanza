## LISTAS ENLAZADAS 

- simples 
- dobles
- circulares 

-> en este caso no hay desplazamientos sino que hacemos MOVIMIENTO DE DIRECCIONES ( o sea movemos las flechas, las referencias )    
-> creacion de nodos independientes -> nuevo objeto -> queremos el contenido y la direccion y la direccion de un nodo siguiente    
-> nodo -> objeto -> contiene -> contenido  
                              -> apuntador a un siguiente nodo   
-> la cantidad de nodos esta predeterminada por cantidad de memoria no por posicion (diferencia con la lsita sencuencial)   
-> nosotros como usuarios le determinamos la direccion del nodo 
-> el ultimo elemento de la lista enlazada SIEMPRE esta dirigida a None (Que quiere decir?-> None no apunta a ninguna direccion)

```python
class Nodo():
    # constructor
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None
        
    def __str__(self):
        # para devolver una cadena 
        siguiente = self.siguiente.dato if self.siguiente else None
        return f"[{self.dato} | {siguiente}]"
    
    #self.siguiente -> me muestra el nodo 2
    #self.siguiente.dato -> me muestra el dato del nodo 2
    
    # supongamos:
    # nodo1 = 8
    # nodo2 = 9
    # nodo1.dato = 8
    # nodo1.siguiente.dato = 9
    
    
# if __name__ == "__main__":
#     nodo1 = Nodo(1) # aca dato = 1
#     nodo2 = Nodo(2)
#     nodo3 = Nodo(3)
#     print(nodo1)  
#     print(nodo2)
#     nodo1.siguiente = nodo2
#     nodo2.siguiente = nodo3
#     print(nodo1)
#     print(nodo2)
#     print(nodo3)
    
#     actual = nodo1
#     while actual:
#         print(actual.dato)
#         actual = actual.siguiente


class ListaEnlazada(): 
    
    # Almacena la direccion del primer nodo de la lista
    
    # Que pasa si la lista no tiene almacenada ningun nodo? Porque todavia no se agrego nada
    
    # dirigida a None -> direccionada a ninguna direccion de memoria -> lista vacia
    
    # cuando decimos por ejemplo:
        # nodo1 = 8 -> ahi ya la direccionamos a un valor 

    # si quiero agregar un elemento a la lista, lo voy a agregar al momento en el que encuentre a None
    # para eso, necesito tener un puntero que me indique donde estoy en la lista
    # ese puntero se llama "actual"
    # al principio, "actual" va a ser igual a "inicio"
    # inicio es el primer nodo de la lista
    # si inicio es None, significa que la lista esta vacia
    # si inicio no es None, significa que la lista tiene al menos un nodo
    # en este caso, voy a recorrer la lista hasta encontrar el primer nodo que tenga "siguiente" igual a None
    # una vez que encuentre ese nodo, voy a agregar el nuevo nodo en esa posicion
    
    # constructor
    def __init__(self):
        self.inicio = None
    
    def es_vacia(self):
        return self.inicio is None
        # esto es lo mismo que poner: return self.inicio == None
    
    def agregar_al_final(self,nodo):
        if self.es_vacia():
            self.inicio = nodo
        
        else:
            actual = self.inicio
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nodo # aca estoy en el ultimo nodo, que es el que tiene siguiente = None, y le asigno el nuevo nodo y pueda seguir el recorrido
        
    def agregar_al_inicio(self,nodo):
        if self.es_vacia():
            self.inicio = nodo
        else:
            nodo.siguiente = self.inicio    # el nuevo nodo apunta al inicio actual
            self.inicio = nodo              # el inicio ahora es el nuevo nodo
    
    
    # def __str__(self):  el str devuelve una cadena, entonces deberiamos recorrer la lista y devolver una cadena con todos los nodos
    def mostrar(self):
        if self.es_vacia():
            print("La lista esta vacia")
        
        else:
            actual = self.inicio            # variable local que me permite recorrer la lista
            while actual:
                print(actual)               # esto llama al str de la clase nodo
                actual = actual.siguiente
    
    
    # forma de hacerlo con str
    def __str__(self):
        if self.es_vacia():
            return "La lista esta vacia"
        
        else:
            actual = self.inicio
            nodos = []
            while actual:
                nodos.append(str(actual))    # esto llama al str de la clase nodo
                actual = actual.siguiente
            return " -> ".join(nodos)        # une todos los elementos de la lista con " -> " entre ellos

if __name__ == "__main__":
    lista = ListaEnlazada()
    print(lista) # este print llama al str de la clase lista 
    # sino deberia hacer lista.mostrar()
    nodo1 = Nodo(1)
    nodo2 = Nodo(2)
    nodo3 = Nodo(3)
    lista.agregar_al_final(nodo1)
    print(lista)
    lista.agregar_al_final(nodo2)
    print(lista)
    lista.agregar_al_final(nodo3)
    print(lista)
    nodo0 = Nodo(0)
    lista.agregar_al_inicio(nodo0)
    print(lista)
        
```
