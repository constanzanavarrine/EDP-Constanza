### KWARGS



-> me permite guardar una cantidad inespecifica de datos para luego trabajar con diccionarios   
-> se ajusta a la cantidad de datos     
-> como manipular la informacion sin necesidad de que la variable pasada por parametro sea un diccionario       

```python
estudiantes_nota = {
    'fede': [8, 9],
    'juan': [7, 6],
    'carlos': [4, 3]
}

def ver_notas(estudiantes_notas):
    for estudiante, notas in estudiantes_notas.items():
        print(f' El estudiante {estudiante}, tiene como notas {notas}')

ver_notas(estudiantes_nota)

```

con kwargs lo que hacemos es por ejemplo en ver_notas, no enviar un diccionario, sino variables que luego estaran representadas en diccionarios         

-> valor -> asociado a un dato

```python


def ver_notas1(**kwargs):
    for estudiante, nota in kwargs.items():
        print(f'El estudiante {estudiante}, tiene como notas {nota}')

def promedio_estudiante(notas):
    nota = sum(notas)
    return nota/len(notas) if len(notas) > 0 else 0

def nota_cursada(**kwargs):
    notas = list(map(promedio_estudiante, kwargs.values()))
    suma_notas = sum(notas)
    return suma_notas / len(kwargs) if len(notas) > 0 else 0



ver_notas1(fede = [8,9], juan = [7,6], carlos = [4,3], chiara = [8,10])
print(nota_cursada(fede = [8,9], juan = [7,6], carlos = [4,3], chiara = [8,10]))
```

### CONJUNTOS

-> estructura desordenada   
-> no tiene elementos repetidos     
-> no puedo modificar un dato de la estructura      
-> si puedo agregar datos y eliminar datos      
-> iterable     
-> no es secuencial     
-> no es indexada   
-> no puedo acceder a los datos por indice  
-> adicionalmente es la forma de simular el comportamiento de los conjuntos     
-> puede contener diferentes tipos de datos     
-> son mutables     
-> internamente no se pueden modificar los datos que tienen     
-> no nos muestra necesariamente los datos en el mismo orden        

```python
# Constructor 
# Set

conjunto1 = set() # -> solo acepta un unico argumento
print(type(conjunto1))
print(conjunto1)

# a traves de llaves 
#conjunto2 = {} -> mal no puede ser con llaves vacias porq sino se crea un diccionario
conjunto2 = (1,)
print(type(conjunto2))
print(conjunto2)

# Elementos 
#conjunto2 = {1, 2, 3, 'Chiara', 'Juan', [1, 2, 4]} -> la lista no puede ir dentro de un conjunto, porque no lo podemos recorrer
conjunto2 = {1, 2, 3, 'Chiara', 'Juan'}
conjunto2 = {1, 2, 3, 'Chiara', 'Juan', (1, 2, 4)} 
print(conjunto2)
```

```python
# sin elementos repetidos 
conjunto3 = set((1, 2, 3, 2, 3, 4, 5, 6, 6, 7, 9))
print(conjunto3)
```

```python
lista = [1, 2, 3, 2, 3, 4, 5, 6, 6, 7, 9]
lista = list(set(lista))
print('lista',lista)
```

### Moverse por un conjunto


```python
conjunto = set()
for elemento in conjunto:
    print(elemento)
```

### Agregar dato a un conjunto

```python
conjunto = set()
conjunto.add('dato')
print(conjunto)
```

### Agregar mas de un elemento 
#### Elementos independientes

```python
conjunto = set()
conjunto.update(('hola',10))
conjunto.update('Miranda')
print(conjunto)

```

### Eliminar datos de un conjunto

```python
# eliminar pop()
# eliminar remove()
# discard
conjunto = set()
conjunto.update(('hola',10))
conjunto.update('Miranda')
print(conjunto)

# eliminar pop() -> elimina en forma aleatoria 
conjunto.pop()
print(conjunto)

# discard
conjunto.discard('saludo')
print(conjunto)

# clear
conjunto.clear()
print(conjunto)
```

### RECURSIVIDAD 

- Cuando la implementacion a traves de ciclos for se vuelve compleja usamos recursion 
- Una funcion se llama a ella misma
- Buscar un caso base -> condicion que hace que el proceso termine 
- Debe existi al menos UN caso base 
- Cada llamado a la funcion agrega cada entrada a una pila de llamadas  
- Factorial, Mostrar lista, Sumas lista, Fibonacci, Multiplicar, Division 

```python
def saludo():
    print('hola')
    saludo() 
    # entramos en un ciclo sin fin
    # no hay nada que me detenga el llamado 
    # tenemos que definir algo para que no se convierta en una funcion infinita

saludo()
```

```python
# Este caso no tiene caso base entonces mas alla de que 
# la funcion se llama a ella misma, no es recursiva porque 
# no tiene caso base 

def imprimir(x):
    print(x)
    imprimir(x-1)

imprimir(4)
```

```python
def suma(x):
    print(x)
    return x + suma(x-1)

suma(5)
# visualiza     # llama 
# 5             return 5 + suma(5-1)
# 4             return 4 + suma(4-1)
# 3             return 3 + suma(3-1)
```

```python
def imprimir(x):
    if x > 0: 
        print(x)
        imprimir(x-1)

imprimir(4)
# termina cuando llega a los negativos 
```

```python
def producto(n):
    if n == 0:   # caso base que hace que el cico se interrumpa 
        return 1
    else:
        return n * producto(n-1)

print(producto(3))
```

#### Mostrar una lista en forma recursiva

```python
def mostrar_lista(lista, indice): 
    if indice == len(lista):
        return 
    
    print(lista[indice])
    mostrar_lista(lista,indice+1)

mostrar_lista([2,4,1,3],0)
        
```

```python
# sin indice
def mostrar_lista(lista):
    if len(lista) > 0:
        print(lista[0])
        mostrar_lista(lista[1:])

mostrar_lista([2,4,1,3])
```
