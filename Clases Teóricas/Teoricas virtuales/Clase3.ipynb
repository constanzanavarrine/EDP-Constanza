Programacion orientada a objetos

4 caracteristicas principales: 

Abstraccion         → representar el elemento del mundo real lo mas parecido que podamos en el codigo → determinar atributos relevantes al proyecto en trabajo

Herencia            →  reusabilidad →  relacion entre clases →  buscamos crear clases generales y luego extender sus atributos particularizando determinados tipos de objetos
                    →  Persona →  Estudiante (atributos propios)
                    →  como det si hay herencia? si yo sigo un estudiante es una persona (compañia con persona no)

Encapsulamiento     →  el programador pueda acceder o no a la info de la clase dependiendo si los atributos son publicos, privados o protegidos
                    →  ocultar detalles internos de las clases para que no sufran daños externos

Modularidad         →  crear modulos independientes donde hagamos tareas especificas 
                    →  puedo verificar que la biblioteca existe (una vez que lo probé lo puedo empezar a relacionar)

Como creamos una clase?
1. Hacer una plantilla 
2. Nombre de las clases siempre en mayusculas
3. Los parentesis se colocan cuando hay herencia 
4. Toda clase tiene un constructor 


def __int__(self, titulo: str, autor: str, editorial: str, ISBN: str, estado = 'disponible)

# constructor → init 

5. self →  almaceno la info del constructor (nos referimos al objeto) → o sea el self representa el objeto
6. atributos (luego del atributo debemos poner a que tipo de dato corresponde)
7. atributos de instancia siempre van con self → le damos valores propios al objeto en particular 
8. el init siempre lleva self 

parametros pasados a la funcion init:
ej. titulo: str → typehint



```python
class Libro:
    
    # metodo 1 (de instancia porque tiene self)
        # constructor (guarda info en el espacio de memoria)
            # # creacion del objeto a partir de los datos por parametro
    def __init__(self, titulo: str, autor: str, editorial: str, ISBN: str, estado = 'disponible'): # aca lo que hago es construir el objeto 
        self.titulo = titulo                # atributos de instancia → (es igual a lo que llega por parametro)
        self.autor = autor
        self.editorial = editorial
        self.ISBN = ISBN                    # (identificador unico)
        self.estado = estado



    # metodo 2 (de instancia porque tiene self)
        # visualizar la informacion de cada libro
        # es un metodo por defecto de Python para la visualizacion 
    def __str__(self):    
          
        # devuelve siempre una cadena 
        return (f'El libro {self.titulo} es el autor {self.autor} y la editorial es {self.editorial} con ISBN {self.ISBN}')
    


# Llamo a la clase para darle valores a los atributos 

rayuela = Libro('Rayuela', 'Julio Cortazar', 'Argentina', 'NSE123')
    # OBJETO: rayuela
    # Atributos: 'Rayuela', 'Julio Cortazar', 'Argentina', 'NSE123'
    # No pasamos estado porque ya esta determinado arriba como predeterminado (disponible)
Coronel = Libro('El coronel no tiene quien le escriba', ' Gabriel Garcia Marquez', 'Colombia', 'NSE456', 'prestado')

# llamamos 2 veces a la clase Libro → creamos 2 objetos

print(rayuela)
print(Coronel)
    
# si pusieramos toggle en def str entonces cuando imprimo me mostraria 
# la direccion del objeto 
    
    
    
```

```python
class Libro:
    
    def __init__(self, titulo: str, autor: str, editorial: str, ISBN: str, estado = 'disponible'): 
        self.titulo = titulo                
        self.autor = autor
        self.editorial = editorial
        self.ISBN = ISBN                    
        self.estado = estado

    # Caso particular de visualizacion 
    def mostrar(self):    
          
        # devuelve siempre una cadena 
        return (f'El libro {self.titulo} es el autor {self.autor} y la editorial es {self.editorial} con ISBN {self.ISBN}')
    

rayuela = Libro('Rayuela', 'Julio Cortazar', 'Argentina', 'NSE123')
Coronel = Libro('El coronel no tiene quien le escriba', ' Gabriel Garcia Marquez', 'Colombia', 'NSE456', 'prestado')

print(rayuela.mostrar())
print(Coronel.mostrar())
    
```
