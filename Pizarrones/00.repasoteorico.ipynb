## LISTAS SECUENCIALES

#### Caracteristicas
+ Estructura dinamica: mutable. Modificable en el tiempo de ejecucion       
+ Estructura secuencial: los datos internamente van uno despues del otro        
+ Estructura ordenada: los datos siempre estan en la misma posicion y por tanto los puedo recorrer por indice       
+ Estructura acceder: indice (while/range) (contenido for dato in lista)            
+ Datos: almacena todo tipo de datos y no necesariamente son homogenos      
+ Le asigna una capacidad de almacenamiento:        
    - Python busca un nuevo espacio de memoria                 
    - copia todos los elementos iniciales en mi lista       
    - eliminar o liberar el espacio de memoria      
    - sigo operando con la nueva lista 
+ Agregar o eliminar datos de una posicion > movimiento de todos los datos              
    - Mucha informacion     
    - Operaciones son repetitivas            
+ Me permite tener elementos repetidos          
+ Busquedas .. Posiciones / por elemento        

#### Ejemplos

+ Guardar los resultados de cálculo de unos procesos        
+ Lista de reproducción, si la reproducción de la lista es en el orden que fue creada y es secuencial       
+ Muestra el historial      

#### Conclusion 

Excelente en que casos:
+ acceder a la información en el orden que lo agregué       
+ Tamaño de la lista no es variable mas estable     
+ Accede a traves de índice     


#### Problema
+ Agregar / eliminar en forma frecuente     
+ No información repetida       
+ Búsqueda cuando tengo mucha información       


## LISTAS ENLAZADAS SIMPLES

#### Características

+ Estructura Dinámica: Si
+ Estructura Secuencial: Si. La secuencia está dada por las referencias al siguiente
+ Estructura Ordenada: Si
+ Tipos de Datos: Nodos, siempre el mismo tipo
+ Que implementación van a usar?
+ Iteración: Si, adelante
+ Elementos repetidos. Nodos No, la referencia a los nodos debe ser única, pero el dato o
contenido del nodo puede repertirse
+ Agregar/ Eliminar elementos. Más eficiente. Solo cambia referencias


#### Ejemplos
- Lista de tareas, agregando/eliminado elementos en forma dinámica.


#### Conclusión

Excelente en que casos:
- Agregar/eliminar datos en forma muy frecuente
- Trabajando con muchos datos
- Necesito que la estructura maneje solo un tipo de dato


#### Problemas
- Acceder mucho al final
- Acceder por índice no puedo
- Cuando no se manejan grandes volúmenes de Datos y las tareas de agregar y eliminar no es
frecuente
